# 00 Environment and Infrastructure Setup

## Summary
This notebook prepares the foundational cloud infrastructure and authentication setup required for the Agentic AI Financial Intelligence Platform.

It focuses on:
- Google Cloud project setup
- Authentication configuration
- Terraform-based infrastructure provisioning
- Environment variable configuration
- Vertex AI Workbench access

## Requirements / Prerequisites

Before running this notebook, ensure you have:
- A Google Cloud account
- Billing enabled
- Terraform installed
- Google Cloud SDK installed
- Appropriate IAM permissions


## Clone Repository

Clone the repository locally before beginning infrastructure provisioning.


```
git clone https://github.com/<YOUR_USERNAME>/agentic-ai-financial-intelligence-platform.git
cd agentic-ai-financial-intelligence-platform
```

## Authenticate with Google Cloud

Authenticate both the Google Cloud CLI and Application Default Credentials.


```
gcloud auth login
gcloud auth application-default login
```

## Configure Project Context

Replace placeholder values before execution.


```
# Configuration: Replace placeholders with your GCP project details

gcloud config set project <PROJECT_ID>
gcloud auth application-default set-quota-project <PROJECT_ID>
```

## Define Terraform Environment Variables

Replace placeholder values before execution.

Passwords should follow strong password complexity requirements.


```
export TF_VAR_gcp_project_id="<PROJECT_ID>"
export TF_VAR_region="<REGION>"
export TF_VAR_alloydb_password="<ALLOYDB_PASSWORD>"
export TF_VAR_cloud_sql_password="<CLOUD_SQL_PASSWORD>" 
```

## Initialize Terraform

Terraform provisions:
- Networking
- AlloyDB AI
- Spanner Graph
- Vertex AI Workbench
- Cloud Run dependencies
- IAM and security resources


```
cd terraform/

terraform init
terraform plan
terraform apply
```

## Access Vertex AI Workbench

After Terraform completes, copy the generated Workbench proxy URL and open it in your browser.

To avoid unnecessary cloud charges:
- Destroy Terraform-managed resources when no longer needed

```
# Terraform cleanup

terraform destroy
```
